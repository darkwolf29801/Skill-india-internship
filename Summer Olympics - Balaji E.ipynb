```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
summer=pd.read_csv('summer.csv')
s=np.array(summer)
```

# 1. In how many cities Summer Olympics is held so far?

```python
count=0;
for city in summer['City'].unique():
    count+=1
print(count)
```

# 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
data = []

for sport in summer['Sport'].unique():
    data.append([sport , len(summer[summer['Sport']  == sport])])
pd.DataFrame(data[:5],columns = ['Sport','Medal']).sort_values(by='Medal', ascending=False).plot(x = 'Sport', y = 'Medal', kind = 'bar', figsize = (5,5),color='b');
```

# 3. Which sport is having most number of medals so far? (Top 5)

```python
medals=[]
for city in summer['City'].unique():
    count =0 
    for i in summer['City']:
        if i==city:
            count+=1
    medals.append([city , count]);
pd.DataFrame(medals[:5],columns = ['City','Medal']).sort_values(by='Medal', ascending=False).plot(x = 'City', y = 'Medal', kind = 'bar', figsize = (5,5),color='m');
```

# 4. Which player has won most number of medals? (Top 5)

```python
players = []

for player in summer['Athlete'].unique():
    players.append([player , len(summer[summer['Athlete']  == player])]);
pd.DataFrame(players[:5],columns = ['Athlete','Medal']).sort_values(by='Medal', ascending=False).plot(x = 'Athlete', y = 'Medal', kind = 'bar', figsize = (5,5));
```

# 5. Which player has won most number Gold Medals of medals? (Top 5) 

```python
top=[]
for j in summer['Country'].unique():
    count=0
    for i in range(len(summer)):
        if s[i][5]==j and s[i][8]=='Gold':
            count+=1
    top.append([j,count])
pd.DataFrame(top[:5],columns = ['Country','Gold']).sort_values(by='Gold', ascending=False).plot(x = 'Country', y = 'Gold', kind = 'bar', figsize = (5,5),color='g');
    
    
```

# 6. In which year India won first Gold Medal in Summer Olympics?

```python
year=2022
for i in range(len(summer)):
    if s[i][5]=='IND' and s[i][8]=='Gold':
            if int(s[i][0])<int(year):
                year=s[i][0]
print(year)
```

# 7. Which event is most popular in terms on number of players? (Top 5)

```python
events=[]
for i in summer['Event'].unique():
        count=0
        for j in range(len(summer)):
            if i==s[j][7]:
                count+=1
        events.append([i,count])
pd.DataFrame(events[:5],columns = ['Event','Number of players']).sort_values(by='Number of players', ascending=False).plot(x = 'Event', y = 'Number of players', kind = 'bar', figsize = (5,5),color='r');
```

# 8. Which sport is having most female Gold Medalists? (Top 5)

```python
women=[]
for i in summer['Sport'].unique():
        count=0
        for j in range(len(summer)):
            if i==s[j][2] and s[j][6]=='Women' and s[j][8]=='Gold':
                count+=1
        women.append([i,count])
pd.DataFrame(women[:5],columns = ['Sport','Number of female gold medalists']).sort_values(by='Number of female gold medalists', ascending=False).plot(x = 'Sport', y = 'Number of female gold medalists', kind = 'bar', figsize = (5,5),color='y');
```
